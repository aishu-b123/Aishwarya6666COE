```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
print(tips.head())
```

```python
sns.histplot(data=tips,x='total_bill',kde=True,bins=20,color='pink')
plt.title('Distribution of total bills')
plt.show()
```

```python
sns.kdeplot(data=tips,x='total_bill',color='skyblue',shade=True)
plt.title('Distribution of total bills')
plt.show()
```

```python
sns.barplot(data=tips,x='day',y='total_bill',hue='day',palette='coolwarm')
plt.title('Distribution of total bills')
plt.show()
```

```python
sns.barplot(data=tips,x='day',y='total_bill',hue='day',palette='colorblind')
plt.title('Distribution of total bills')
plt.show()
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',palette='YlGnBu')
plt.title('Distribution of total bills')
plt.show()
```

```python
sns.scatterplot(data=tips,y='tip',x='total_bill',hue='day',palette='muted')
plt.title('Scatter plot of Total Bill vs Tip')
plt.show()
```

```python
sns.lineplot(data=tips,y='tip',x='size',color='green',marker='o')
plt.title('Line plot of size vs Tip')
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex',palette='deep')
plt.show()
```

```python
num=tips.select_dtypes(include='number')
corr_matrix=num.corr()
sns.heatmap(corr_matrix,annot=True,cmap='copper',fmt='.2f')
plt.show()
```
