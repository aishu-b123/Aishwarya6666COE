## Probability Distribution ##

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
p=0.6
data=bernoulli.rvs(p,size=1000)
plt.bar([0,1],[sum(data==0),sum(data==1)],color=['red','green'],tick_label=['Failure(0)','success(1)'])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Bernoulli Distribution")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

λ=3
x=np.arange(0,10)
pmf=poisson.pmf(x,λ)

plt.bar(x,pmf,color='blue',alpha=0.7)
plt.xlabel("no of success ")
plt.ylabel('Probability')
plt.title(f"Poisson Distribution (λ={λ})")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mu=70
sigma=10
x=np.linspace(mu-4*sigma,mu+4*sigma,100)
pdf=norm.pdf(x,mu,sigma)
plt.plot(x,pdf,color='red',label=f"N{mu},{sigma}²")
plt.xlabel("Exam scores ")
plt.ylabel('Probability Density')
plt.title(f"Normal Distribution (µ={λ}),(õ={sigma})")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
z=np.linspace(-4,4,1000)
pdf=norm.pdf(z,0,1)
plt.plot(z,pdf,color='blue',label='Standard Normal Distribution')
plt.xlabel("z-score")
plt.ylabel('Probability Density')
plt.title("Standard Normal Distribution")
plt.axvline(x=0,color='red',linestyle='--',label="Mean")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
income_data=np.random.exponential(scale=50,size=1000)
plt.figure(figsize=(12, 6))
sns.histplot(income_data, bins=50,kde=True ,color='r' )
plt.title(' Original Income')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
 
sample_means=[]
n_sam=1000
sam_size=30
 
for _ in range(n_sam):
    sam=np.random.choice(income_data,size=sam_size)
    sample_means.append(np.mean(sam))

plt.figure(figsize=(12, 6))
sns.histplot(sample_means, bins=50,kde=True ,color='b' )
plt.title(' Sampling  Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```
