```python
import numpy as np
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df
df = remove_outliers(df, "price")
df = remove_outliers(df, "powerPS")
```

```python
df['notRepairedDamage'].unique()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[["price", "powerPS"]] = scaler.fit_transform(df[["price", "powerPS"]])
df['price'].head()
```

```python
df["dateCrawled"].dtype
```

```python
df["dateCrawled"] = pd.to_datetime(df["dateCrawled"])
df["yearCrawled"] = df["dateCrawled"].dt.year
df["monthCrawled"] = df["dateCrawled"].dt.month
df["dayCrawled"] = df["dateCrawled"].dt.day
df["weekdayCrawled"] = df["dateCrawled"].dt.weekday
df["hourCrawled"] = df["dateCrawled"].dt.hour
```

```python
df["mileage_per_year"] = df["kilometer"] / (2025 - df["yearOfRegistration"])
df["mileage_per_year"].head()
```

```python
from sklearn.preprocessing import LabelEncoder

df["notRepairedDamage"] = df["notRepairedDamage"].map({"nein": 0, "ja": 1, "unknown": 2})
df["abtest"] = df["abtest"].map({"control": 0, "test": 1})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Average price per brand
brand_prices = df.groupby("brand")["price"].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 5))
sns.barplot(x=brand_prices.index, y=brand_prices.values, palette="viridis")
plt.xticks(rotation=90)
plt.title("Average Car Price by Brand")
plt.xlabel("Brand")
plt.ylabel("Average Price")
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False, drop='first')

# Select categorical columns to encode
categorical_cols = ["vehicleType", "gearbox"]

# Encode categorical columns
encoded_cols = pd.DataFrame(
    encoder.fit_transform(df[categorical_cols]),  
    columns=encoder.get_feature_names_out(categorical_cols)  # Get meaningful column names
)

# Reset the index to match df
encoded_cols.index = df.index  

# Drop original categorical columns
df.drop(columns=categorical_cols, inplace=True)

# Concatenate the encoded features with the original dataframe
df = pd.concat([df, encoded_cols], axis=1)

# Display the updated dataset
print(df.head())
```

```python
df['model'].fillna(df['model'].mode()[0],inplace=True)
```

```python
df["price_per_km"] = df["price"] / df["kilometer"]
df["price_per_km"].replace([np.inf, -np.inf], np.nan, inplace=True)  # Replace infinities
df["price_per_km"].fillna(df["price_per_km"].median(), inplace=True)  # Fill missing values
```
