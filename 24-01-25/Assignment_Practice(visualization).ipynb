```python
import matplotlib.pyplot as plt

Month=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Sales= [2000, 2500, 2400, 2800, 3000, 3500, 3800, 4200, 4600, 5000, 5300, 5500]

plt.plot(Month,Sales, marker='o',color='orange',linewidth=2)
plt.title('Monthly Sales Analysis')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.show()
```

```python
order_quantities = [5,6,7,7,8,5,3,4,4,4,6,7,2,3,4,2]
b=[2,4,6,8,10]
plt.hist(order_quantities, bins=b,edgecolor='black')
plt.title('Order Quantity Distribution')
plt.xlabel('Order Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
ages = [22, 34, 26, 40, 58, 29, 19, 35, 44, 52, 31, 29, 24, 41, 60, 38, 28, 33, 25, 45]
b=[10,20,30,40,50,60,70,80,90]
plt.hist(ages, bins=b, color='magenta',edgecolor='black') 
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
spending = [500, 600, 700, 800, 1200, 900, 450, 1000, 1100, 650, 750, 850, 600, 1300, 1400, 1200, 950, 750, 800, 1000]
plt.scatter(ages, spending,color='yellow',edgecolor='black',s=50)
plt.title('Customer Spending')
plt.xlabel('Age')
plt.ylabel('Spending')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
marks_data = {
    'Math': [85, 78, 92, 88, 95, 60, 72, 80, 35, 90],
    'Science': [90, 85, 78, 88, 92, 81, 80, 83, 75, 93],
    'English': [88, 90, 78, 85, 80, 72, 95, 70, 80, 82],
    'Telugu': [72, 65, 75, 80, 70, 78, 60, 30, 72, 60],
    'Social': [95, 92, 88, 100, 98, 90, 91, 85, 89, 87]
}
marks_df = pd.DataFrame(marks_data)
plt.figure(figsize=(8, 6))
plt.boxplot(marks_df.values,labels=marks_df.columns)
plt.title('Box Plot for Student Marks ')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```
