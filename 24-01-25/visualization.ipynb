```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y,color="pink",linewidth=3,linestyle='dashed',marker='o')
plt.title("Line plot")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.show()
```

create a line plot showing the growth of a savings account over 6 months 
1 x-axis(1-6)

```python
x=[1,2,3,4,5,6]
y=[5000,2000,500,1000,700,1200]
plt.plot(x,y,color='green',linestyle='dashed',marker='o')
plt.xlabel('Months')
plt.ylabel('Savings')
plt.title("Line plot for Savings")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[15,6,12,9,7]
plt.bar(categories,values,color=('magenta'))
plt.xlabel('Categories')
plt.ylabel('values')
plt.title("Line plot for categories")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[15,6,12,9,7]
plt.bar(categories,values,color=('skyblue','pink','orange','purple','lightgreen'),edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('values')
plt.title("Line plot for categories")
plt.show()
```

```python
ages=[20,21,36,27,45,22,23,24,48,25,26,27,28,32,33,34,30,31,25,33,34,35,36,37,38,39,40]
b=[20,25,30,35,40,45,50]
plt.hist(ages,bins=b,color='pink',edgecolor='black')
plt.xlabel('Ages')
plt.ylabel('values')
plt.title("Line plot for Ages")
plt.grid(True)
plt.show()
```

```python
languages=['python','c','java','c++','javascript']
per=[40,20,80,47,8]
plt.pie(per,labels=languages,colors=('pink','yellow','skyblue','lightgreen','orange'),autopct='%1.2f%%')
plt.title("Line plot for Ages")
plt.legend()
plt.show()
```

```python
x=[5,10,15,10,25,12,18,20]
y=[10,20,30,35,20,22,12,15]
plt.scatter(x,y,color='orange',s=50,edgecolor='black')
plt.show()
```
